```python
import gdown
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://drive.google.com/uc?id=1PLppvwAQ-9sRYAZGZt54sjBo5r8uBDDR'
output = 'Coffee_Stores_Data.csv'
gdown.download(url, output, quiet=False)

# Load data into a DataFrame
df = pd.read_csv('/content/Coffee_Stores_Data.csv')

#Data Prep
# Convert 'BusinessDate' to datetime format if not already
df['BusinessDate'] = pd.to_datetime(df['BusinessDate'])

# Extract day of the week and create a new column 'DayOfWeek'
df['DayOfWeek'] = df['BusinessDate'].dt.day_name()

# Display unique values in the 'DayOfWeek' column
print(df['DayOfWeek'].unique())

# Check for missing values
print(df.isnull().sum())

# Display basic statistics
print(df.describe())

# Display information about the dataset
print(df.info())

# Analysis
#Store18
selected_stores = [18]
selected_data = df[df['StoreID'].isin(selected_stores)]
# Enhance the readability of box plots
plt.figure(figsize=(18, 14))

# Box plots for 27 products
plt.subplot(2, 2, 1)
sns.boxplot(x='Description', y='SoldQuantity', data=selected_data)
plt.title('Box Plots for 27 Products', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sold Quantity', fontsize=12)

# Box plots for inventory patterns
plt.subplot(2, 2, 2)
sns.boxplot(x='Description', y='EndQuantity', data=selected_data)
plt.title('Box Plots for Inventory Patterns', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('End Quantity (Inventory)', fontsize=12)

# Box plots for stock out patterns
plt.subplot(2, 2, 3)
sns.boxplot(x='Description', y='StockedOut', data=selected_data)
plt.title('Box Plots for Stock Out Patterns', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Stocked Out', fontsize=12)

# Box plots for missed sales
plt.subplot(2, 2, 4)
sns.boxplot(x='Description', y='MissedSales', data=selected_data)
plt.title('Box Plots for Missed Sales', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Missed Sales', fontsize=12)

plt.tight_layout()
plt.show()


#Store 117
selected_stores = [117]
selected_data = df[df['StoreID'].isin(selected_stores)]
# Enhance the readability of box plots
plt.figure(figsize=(18, 14))

# Box plots for 27 products
plt.subplot(2, 2, 1)
sns.boxplot(x='Description', y='SoldQuantity', data=selected_data)
plt.title('Box Plots for 27 Products', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sold Quantity', fontsize=12)

# Box plots for inventory patterns
plt.subplot(2, 2, 2)
sns.boxplot(x='Description', y='EndQuantity', data=selected_data)
plt.title('Box Plots for Inventory Patterns', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('End Quantity (Inventory)', fontsize=12)

# Box plots for stock out patterns
plt.subplot(2, 2, 3)
sns.boxplot(x='Description', y='StockedOut', data=selected_data)
plt.title('Box Plots for Stock Out Patterns', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Stocked Out', fontsize=12)

# Box plots for missed sales
plt.subplot(2, 2, 4)
sns.boxplot(x='Description', y='MissedSales', data=selected_data)
plt.title('Box Plots for Missed Sales', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Missed Sales', fontsize=12)

plt.tight_layout()
plt.show()

#Store 332
selected_stores = [332]
selected_data = df[df['StoreID'].isin(selected_stores)]
# Enhance the readability of box plots
plt.figure(figsize=(18, 14))

# Box plots for 27 products
plt.subplot(2, 2, 1)
sns.boxplot(x='Description', y='SoldQuantity', data=selected_data)
plt.title('Box Plots for 27 Products', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sold Quantity', fontsize=12)

# Box plots for inventory patterns
plt.subplot(2, 2, 2)
sns.boxplot(x='Description', y='EndQuantity', data=selected_data)
plt.title('Box Plots for Inventory Patterns', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('End Quantity (Inventory)', fontsize=12)

# Box plots for stock out patterns
plt.subplot(2, 2, 3)
sns.boxplot(x='Description', y='StockedOut', data=selected_data)
plt.title('Box Plots for Stock Out Patterns', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Stocked Out', fontsize=12)

# Box plots for missed sales
plt.subplot(2, 2, 4)
sns.boxplot(x='Description', y='MissedSales', data=selected_data)
plt.title('Box Plots for Missed Sales', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Missed Sales', fontsize=12)

plt.tight_layout()
plt.show()


#Task 2: Best and worst sellers
df = pd.read_csv('/content/Coffee_Stores_Data.csv')


average_selling_price = 3
df['InventoryCost'] = df['SoldQuantity'] * average_selling_price

# Filter data for stores 17, 188, and 332
selected_stores = [17, 188, 332]
selected_data = df[df['StoreID'].isin(selected_stores)]

# Calculate total inventory cost per product
total_inventory_cost_per_product = selected_data.groupby('Description')['InventoryCost'].sum()

# Set up figures and axes for each store
fig, axes = plt.subplots(nrows=len(selected_stores), ncols=2, figsize=(15, 7 * len(selected_stores)))

for i, store_id in enumerate(selected_stores):
    # Identify top and bottom 25% products based on inventory cost for each store
    top_25_products_store = total_inventory_cost_per_product[selected_data[selected_data['StoreID'] == store_id]['Description']].sort_values(ascending=False).head(int(len(total_inventory_cost_per_product) * 0.25))
    bottom_25_products_store = total_inventory_cost_per_product[selected_data[selected_data['StoreID'] == store_id]['Description']].sort_values().head(int(len(total_inventory_cost_per_product) * 0.25))

    # Filter data for top and bottom products for each store
    top_products_data_store = selected_data[(selected_data['StoreID'] == store_id) & (selected_data['Description'].isin(top_25_products_store.index))]
    bottom_products_data_store = selected_data[(selected_data['StoreID'] == store_id) & (selected_data['Description'].isin(bottom_25_products_store.index))]

    # Check if the DataFrame is not empty before plotting
    if not top_products_data_store.empty:
        # Plotting for top 25% products based on inventory cost
        sns.barplot(x='Description', y='InventoryCost', data=top_products_data_store, ax=axes[i, 0], color='green')
        axes[i, 0].set_title(f'Store {store_id} - Top 25% Inventory Cost')
        axes[i, 0].set_xlabel('Product')
        plt.xticks(rotation=45, ha='right')
        axes[i, 0].set_ylabel('Inventory Cost')

    # Check if the DataFrame is not empty before plotting
    if not bottom_products_data_store.empty:
        # Plotting for bottom 25% products based on inventory cost
        sns.barplot(x='Description', y='InventoryCost', data=bottom_products_data_store, ax=axes[i, 1], color='orange')
        axes[i, 1].set_title(f'Store {store_id} - Bottom 25% Inventory Cost')
        axes[i, 1].set_xlabel('Product')
        plt.xticks(rotation=45, ha='right')
        axes[i, 1].set_ylabel('Inventory Cost')

plt.tight_layout()
plt.show()

selected_stores = [17, 117, 332]
selected_data = df[df['StoreID'].isin(selected_stores)]

# Calculate total sales per product
total_sales_per_product = selected_data.groupby('Description')['SoldQuantity'].sum()

# Set up figures and axes for each store
fig, axes = plt.subplots(nrows=len(selected_stores), ncols=2, figsize=(15, 7 * len(selected_stores)))

for i, store_id in enumerate(selected_stores):
    # Identify top and bottom 25% products for each store
    top_25_products_store = total_sales_per_product[selected_data[selected_data['StoreID'] == store_id]['Description']].sort_values(ascending=False).head(int(len(total_sales_per_product) * 0.25))
    bottom_25_products_store = total_sales_per_product[selected_data[selected_data['StoreID'] == store_id]['Description']].sort_values().head(int(len(total_sales_per_product) * 0.25))

    # Filter data for top and bottom products for each store
    top_products_data_store = selected_data[(selected_data['StoreID'] == store_id) & (selected_data['Description'].isin(top_25_products_store.index))]
    bottom_products_data_store = selected_data[(selected_data['StoreID'] == store_id) & (selected_data['Description'].isin(bottom_25_products_store.index))]

    # Check if the DataFrame is not empty before plotting
    if not top_products_data_store.empty:
        # Plotting for top 25% best sellers
        sns.barplot(x='Description', y='SoldQuantity', data=top_products_data_store, ax=axes[i, 0], color='skyblue')
        axes[i, 0].set_title(f'Store {store_id} - Top 25% Best Sellers')
        axes[i, 0].set_xlabel('Product')
        plt.xticks(rotation=45, ha='right')
        axes[i, 0].set_ylabel('Sold Quantity')

    # Check if the DataFrame is not empty before plotting
    if not bottom_products_data_store.empty:
        # Plotting for bottom 25% worst sellers
        sns.barplot(x='Description', y='SoldQuantity', data=bottom_products_data_store, ax=axes[i, 1], color='salmon')
        axes[i, 1].set_title(f'Store {store_id} - Bottom 25% Worst Sellers')
        axes[i, 1].set_xlabel('Product')
        plt.xticks(rotation=45, ha='right')
        axes[i, 1].set_ylabel('Sold Quantity')

plt.tight_layout()
plt.show()
```
